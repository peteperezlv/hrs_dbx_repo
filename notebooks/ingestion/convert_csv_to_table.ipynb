```python
# Convert CSV file to Delta table

csv_path = "/Volumes/dev_catalog/brz_landing_hrs/hrs_vol/randhrs1992_2022v1.csv"
table_name = "dev_catalog.brz_raw_hrs.randhrs1992_2022v1"

# Read CSV file
df = spark.read.csv(
    csv_path,
    header=True,
    inferSchema=True
)

print(f"Read {df.count():,} records from CSV")
print(f"Schema:")
df.printSchema()

# Write to Delta table
df.write.mode("overwrite").saveAsTable(table_name)

print(f"\nSuccessfully created table: {table_name}")
```
