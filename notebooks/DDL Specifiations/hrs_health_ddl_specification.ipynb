## 6. Logical Data Model

Parent Entity: Parent Table
Child Entity: TABLE_NAME
Relationship: PARENT_TABLE 1 to Many TABLE_NAME
Business Rules
- Each TABLE_NAME record belongs to exactly one respondent. 
- Every TABLE_NAME record references a valid respondent. 
- A PARENT_TABLE may have zero or one TABLE_NAME record. 
- The relationship is enforced logically by the ETL process. If physical foreign key enforcement is unsupported, preserve the relationship through ETL validation and metadata.

## 9. Data Type Standards
- Measurement values shall be stored as STRING because:
- Preserve original source values 
- Preserve missing-value codes 
- Prevent scientific notation 
- Preserve leading zeros where applicable 
- Simplify ETL processing 
- Maintain consistency across Silver tables 

No data cleansing shall occur during table creation.
