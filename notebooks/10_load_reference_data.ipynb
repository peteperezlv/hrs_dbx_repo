```python
# Load the RESPONDENT_TYPE

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

def load_respondent_type():

    data = [
        (1, "January 2026", "2026-01-01", "2026-03-31", True),
        (2, "April 2026", "2026-04-01", "2026-06-30", True),
        (3, "July 2026", "2026-07-01", "2026-09-30", False)
    ]

    columns = [
        "CohortID",
        "CohortName",
        "StartDate",
        "EndDate",
        "Active"
    ]

    df = spark.createDataFrame(data, columns)

    df.write.mode("append").saveAsTable("silver.cohort")
```
