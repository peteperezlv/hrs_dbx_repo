### Use Case: BMI and Age Descriptive Statistics w/RAND HRS Longitudinal Data

# Use Case ID: UC-DA-001

## Use Case Name

Analyze BMI and Age Distribution of a Sample Population Using the RAND HRS Longitudinal Dataset

## Business Objective

Enable researchers, students, and analysts to use the RAND Health and Retirement Study (HRS) Longitudinal Dataset to perform descriptive statistical analysis on Body Mass Index (BMI) and Age for a selected sample population. The analysis will provide insights into the central tendency and variability of these health measures and support comparisons across race and ethnicity.

## Actors
- Data Analyst
- Public Health Researcher
- Epidemiologist
- Student

## Supporting Systems
- Databricks
- RAND HRS Longitudinal Dataset
- Delta Lake
- SQL Warehouse (optional)
- Python (Pandas, NumPy, Matplotlib)

## Trigger
The analyst selects a population sample from the RAND HRS dataset and requests descriptive statistical analysis.

## Preconditions
- The RAND HRS Longitudinal Data set has been obtained and is accessible to the user.- 
- RAND HRS data has been loaded into Databricks.
- Demographic variables have been standardized.
- BMI and Age variables are available for the selected survey wave(s).
- Missing or invalid values have been identified.
- The user has access to statistical software capable of performing descriptive statistics and data visualization.
- The dataset includes valid, non-null values for BMI, Age, and Race/Ethnicity variables.
- The user has defined the sample population (e.g., filtered by wave/year, demographic criteria, or study cohort).
- The user has basic knowledge of statistical concepts and software navigation.

## Postconditions
The analyst receives:
- Measures of central tendency
- Measures of dispersion statistics for BMI and Age are calculated and documented.
- A comparative interpretation of the strengths/weaknesses of each measure is produced.
- A visual chart displaying mean BMI by race/ethnicity is generated.
- Findings are available for reporting, further analysis, or decision-making.
- Summary report

## Business Rules
- BMI values should be validated against realistic physiological ranges (e.g., 10–80) before analysis.
- Age values should be validated against the RAND HRS study population range (typically 50+ years).
- Race/Ethnicity categories must follow RAND HRS coding conventions for consistency.
- All statistical calculations must use complete-case analysis unless imputation is explicitly specified.

## Functional Requirements Addressed
- FR-1 Select Population
    - The user shall be able to define a sample population using one or more filters, including:
        - Survey wave
        - Age range
        - Gender
        - Race/Ethnicity
        - Cohort
        - Marital status
        - Geographic region (if available)

    - Example
        - Wave = 2020
        - Age >= 50
        - Race = Hispanic
- FR-2: System/user shall calculate mode, median, and mean for BMI.
- FR-3: System/user shall calculate mode, median, and mean for Age.
- FR-4: System/user shall calculate range, variance, and standard deviation for BMI and Age.
- FR-5: System/user shall support comparative interpretation of measures' strengths/weaknesses.
- FR-6: System/user shall produce a graph of mean BMI segmented by race/ethnicity.
- FR-7: Compare Statistical Measures - The user/system shall explain the strengths and weaknesses of each statistic.

| Measure | Strengths | Weaknesses |
|---------|-----------|------------|
| Mean | Uses all observations | Sensitive to outliers |
| Median | Resistant to outliers | Does not use every value |
| Mode | Easy to understand | May not exist or may have multiple values |
| Range | Very easy to calculate | Uses only two observations |
| Variance | Measures total variability | Difficult to interpret due to squared units |
| Standard Deviation | Widely used and easy to interpret | Sensitive to outliers |

## Non-Functional Requirements
- Accuracy: Statistical calculations must match validated formulas (e.g., population vs. sample variance distinctions).
- Usability: Output should be clearly labeled and interpretable by users with basic statistical literacy.
- Performance: Calculations should complete within acceptable time even for large HRS sample sizes (10,000+ records).

## Sample Interpretation Notes (Expected Analytical Output)
- Mean vs. Median for BMI: If BMI distribution is right-skewed (common in health data due to obesity outliers), the mean will be higher than the median — median may better represent "typical" BMI.
- Mode limitations: BMI, being continuous, may show no meaningful mode or multiple modes unless binned into categories (e.g., underweight, normal, overweight, obese).
- Age distribution: Likely more symmetric in HRS given age-eligibility criteria (50+), so mean and median may be closer.
- Standard Deviation vs. Variance: Standard deviation is preferred for interpretability (same units as BMI/Age), while variance is more useful mathematically (e.g., in ANOVA).
- Range limitation: Sensitive to extreme outliers; does not describe distribution shape.
