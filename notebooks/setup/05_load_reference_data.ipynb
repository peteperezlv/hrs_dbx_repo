```python
# Load the RESPONDENT_TYPE

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

def load_respondent_type():

    data = [
        ("January 2026", "2026-01-01", "2026-03-31", True),
        ("April 2026", "2026-04-01", "2026-06-30", True),
        ("July 2026", "2026-07-01", "2026-09-30", False)
    ]

    columns = [
        "CohortName",
        "StartDate",
        "EndDate",
        "Active"
    ]

    df = spark.createDataFrame(data, columns)
    df.write.mode("append").saveAsTable("dev_catalog.slv_cdm_hrs.cohort")

# Call the function to actually load the data
load_respondent_type()
```

```python
# Load the COHORT

from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, DateType, BooleanType

spark = SparkSession.builder.getOrCreate()

def load_cohort():

    data = [
        ("January 2026", None, "2026-01-01", "2026-03-31", True),
        ("April 2026", None, "2026-04-01", "2026-06-30", True),
        ("July 2026", None, "2026-07-01", "2026-09-30", False)
    ]

    schema = StructType([
        StructField("cohort_name", StringType(), True),
        StructField("cohort_description", StringType(), True),
        StructField("start_date", StringType(), True),
        StructField("end_date", StringType(), True),
        StructField("active", BooleanType(), True)
    ])

    df = spark.createDataFrame(data, schema)
    print('at df write mode')
    df.write.mode("append").saveAsTable("dev_catalog.slv_cdm_hrs.cohort")

# Call the function to actually load the data
load_cohort()
```
